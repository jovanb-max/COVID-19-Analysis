```python
#Reports(data sets) stored in the data frames(df), the raw file is read from data depository using csv text file
dfcountry = pd.read_csv("https://raw.githubusercontent.com/CSSEGISandData/COVID-19/web-data/data/cases_country.csv")
dfconfirmed = pd.read_csv("https://raw.githubusercontent.com/CSSEGISandData/COVID-19/web-data/data/cases_country.csv")
dfrecovered = pd.read_csv("https://raw.githubusercontent.com/CSSEGISandData/COVID-19/web-data/data/cases_country.csv")
dfdeaths = pd.read_csv("https://raw.githubusercontent.com/CSSEGISandData/COVID-19/web-data/data/cases_country.csv")
```

```python
#Data cleaning, using rename function for renaming the columns from the report
dfcountry = dfcountry.rename(columns={'Country_Region' : 'Country', 'Long_' : 'Long', 'Mortality_Rate' : 'Mortality Rate', 'Cases_28_Days': 'Cases in last 28 days', 'Deaths_28_Days': 'Deaths in last 28 Days'})
dfconfirmed = dfconfirmed.rename(columns={'Country_Region' : 'Country', 'Long_' : 'Long', 'Mortality_Rate' : 'Mortality Rate', 'Cases_28_Days': 'Cases in last 28 days', 'Deaths_28_Days': 'Deaths in last 28 Days'})
dfrecovered = dfrecovered.rename(columns={'Country_Region' : 'Country', 'Long_' : 'Long', 'Mortality_Rate' : 'Mortality Rate', 'Cases_28_Days': 'Cases in last 28 days', 'Deaths_28_Days': 'Deaths in last 28 Days'})
dfdeaths = dfdeaths.rename(columns={'Country_Region' : 'Country', 'Long_' : 'Long', 'Mortality_Rate' : 'Mortality Rate', 'Cases_28_Days': 'Cases in last 28 days', 'Deaths_28_Days': 'Deaths in last 28 Days'})
```

```python
#using the sort value function from confirmed column to see the top 10 countries with most confirmed cases
#from the top being most infected to less infected
dfcountry_sorted = dfcountry.sort_values('Confirmed', ascending=False).head(10)
```

```python
#using the style attribute of the dfcountry data frame to highlight country confirmed deaths and active cases by writing a function to define the styling rules
fig = go.FigureWidget( layout=go.Layout() )
def highlight_col(x):
    a = 'background-color: red'
    b = 'background-color: yellow'
    c = 'background-color: blue'
    d = 'background-color: green'
    df1 = pd.DataFrame('', index=x.index, columns=x.columns)
    df1.iloc[:, 4] = b
    df1.iloc[:, 5] = a
    df1.iloc[:, 6] = c
    df1.iloc[:, 0] = d

    return df1
#using the interact function automatically creates user interface controls for exploring data interactively
#and runs the IPythonâ€™s widgets.

def show_latest_cases(n):
    n = int(n)
    return dfcountry.sort_values('confirmed', ascending= False).head(n).style.apply(highlight_col, axis=None)

interact(show_latest_cases, n='5')

ipywLayout = widgets.Layout(border='solid 2px green')
ipywLayout.display='none'
widgets.VBox([fig], layout=ipywLayout)
sorted_dfcountry = dfcountry.sort_values('confirmed', ascending= False)
```

```python
#Data visualisation, making a graph to look at the worst hit countries and how many confirmed cases are there
#by writing a function to plot the sorted country data frames to look at the worst hit countries
from google.colab import output
output.enable_custom_widget_manager()
def show_latest_cases(n):
    n = int(n)
    fig = px.scatter(dfcountry_sorted.head(n), x='Country', y='Confirmed', size='Confirmed', color='Country', hover_name='Country', size_max=60)
    fig.show()
    return fig


#adding the interaction so that the users can interactively explore the data

interact(show_latest_cases, n='5')



```

```python
#creating the world map object from the folium library
world_map = folium.Map(location=[11,0], tiles="cartodbpositron", zoom_start=2, max_zoom = 6, min_zoom = 2)

#plotting the data for all confirmed data frame
for i in range(len(dfconfirmed)):
    folium.Circle(
        location=[dfconfirmed.iloc[i]['lat'], dfconfirmed.iloc[i]['long']],
        fill=True,
        radius=(int((np.log(dfconfirmed.iloc[i, 1]+1.00001)))+0.2)*25000,
        fill_color='red',
        color = 'red',
        tooltip = "<div style='margin: 0; background-color: black; color: white;'>"+
                    "<h4 style='text-align:center;font-weight: bold'>"+dfconfirmed.iloc[i]['country'] + "</h4>"
                    "<hr style='margin:10px;color: white;'>"+
                    "<ul style='color: white;;list-style-type:circle;align-item:left;padding-left:20px;padding-right:20px'>"+
                        "<li>Confirmed: "+str(dfconfirmed.iloc[i,1])+"</li>"+
                        "<li>Deaths:   "+str(dfconfirmed.iloc[i,2])+"</li>"+
                        "<li>Death Rate: "+ str(np.round(dfconfirmed.iloc[i,2]/(dfconfirmed.iloc[i,1]+1.00001)*100,2))+ "</li>"+
                    "</ul></div>",
        ).add_to(world_map)


world_map

```
